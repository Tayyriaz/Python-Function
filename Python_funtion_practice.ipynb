```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('Dummy_Sales_Data_v1.csv')

# Replace values in the 'Quantity' column where the value is less than 0 with 0
df['Quantity'] = df['Quantity'].where(df['Quantity'] >= 0)

# Replace values in the 'UnitPrice(USD)' column where the value is less than 0 with 0
df['UnitPrice(USD)'] = df['UnitPrice(USD)'].where(df['UnitPrice(USD)'] >= 0)

# Print the modified DataFrame
print(df)
```
